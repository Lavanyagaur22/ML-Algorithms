```python
# ============================== loading libraries ===========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from collections import Counter
from sklearn.metrics import accuracy_score
# =============================================================================================
```

```python
# Part I
# ============================== data preprocessing ===========================================

# define column names
names = ['x', 'y', 'class']

#loading the training data
# Refer to dataset here: https://www.kaggle.com/deepthiar/toydatasets?source=post_page-----32820d72c6b6----------------------
df = pd.read_csv('/Users/lavanya/ML_GithubWorks/MLAlgo/demo_data/3.concertriccir2.csv', header = None, names = names)
print(df.head())
print(df.tail())

#create design matrix X and target vector y
X = np.array(df.iloc[:, 0:2])  # end index is exclusive
y = np.array(df['class']) # showing you two ways of indexing a pandas df
```

## Simple Cross Validation

```python
# split the data set into train and test

# random_state-> https://stackoverflow.com/a/53754816/10189663

# In case if you specify random_state=None, then whenever samples are selected, 
# every time a different set of points is selected.
# Whereas if an integer value is set to random_state, then whatever 
# set of points are selected in a sample, every time whenever you pick a sample,
# the same set of points come into the sample.

# We can train many models but we need to figure out which model is giving good results and will give better results in production.
# This process of selection of a model is called model selection. 
# Cross-validation is one type of model selection methods.

#X_1-> X train

X_1, X_test, y_1, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# split the train data set into cross validation train and cross validation test
X_tr, X_cv, y_tr, y_cv = train_test_split(X_1, y_1, test_size=0.3)

for i in range (1,30,2):
  # instantiate learning model (k = 30)
    knn = KNeighborsClassifier(n_neighbors=i)

    # fitting the model on crossvalidation train
    knn.fit(X_tr, y_tr)

    # predict the response on the crossvalidation train
    pred = knn.predict(X_cv)

    # evaluate CV accuracy
    acc = accuracy_score(y_cv, pred, normalize=True) * float(100)
    print('\nCV accuracy for k = %d is %d%%' % (i, acc))
    
knn = KNeighborsClassifier(1)
knn.fit(X_tr,y_tr)
pred = knn.predict(X_test)
acc = accuracy_score(y_test, pred, normalize=True) * float(100)
print('\n****Test accuracy for k = 1 is %d%%' % (acc))
```

## 10 fold cross validation

```python
# creating odd list of K for KNN
myList = list(range(0,50))
neighbors = list(filter(lambda x: x % 2 != 0, myList))

#empty list that will hold cv scores
cv_scores = []

#perform 10 fold cv
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_tr, y_tr, cv=3, scoring='accuracy')
    cv_scores.append(scores.mean())

# changing to misclassification error
MSE = [1 - x for x in cv_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print('\nThe optimal number of neighbors is %d.' % optimal_k)

# plot misclassification error vs k 
plt.plot(neighbors, MSE)

for xy in zip(neighbors, np.round(MSE,3)):
    plt.annotate('(%s, %s)' % xy, xy=xy, textcoords='data')

plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()

print("the misclassification error for each k value is : ", np.round(MSE,3))
```

```python
# ============================== KNN with k = optimal_k ===============================================
# instantiate learning model k = optimal_k
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)

# fitting the model
knn_optimal.fit(X_tr, y_tr)

# predict the response
pred = knn_optimal.predict(X_test)

# evaluate accuracy
acc = accuracy_score(y_test, pred) * 100
print('\nThe accuracy of the knn classifier for k = %d is %f%%' % (optimal_k, acc))
```
